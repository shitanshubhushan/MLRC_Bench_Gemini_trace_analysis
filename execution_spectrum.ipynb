```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Replace with your actual JSON file path
json_file_path = "capability_levels.json"
```

```python
with open(json_file_path, 'r') as f:
    data = json.load(f)

# Initialize dictionaries to store max levels for each model-task combination
models = set()
tasks = list(data.keys())
model_task_levels = {}

# Extract models and their maximum levels for each task
for task, task_data in data.items():
    for model, model_data in task_data.items():
        models.add(model)
        if model not in model_task_levels:
            model_task_levels[model] = {}
        model_task_levels[model][task] = max(model_data['levels'])

# Create DataFrame
df = pd.DataFrame(index=list(models), columns=tasks)

# Fill the DataFrame with max levels
for model in models:
    for task in tasks:
        df.loc[model, task] = model_task_levels[model].get(task, None)
```

```python
df = df.reset_index()
```
